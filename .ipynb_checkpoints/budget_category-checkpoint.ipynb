# Category Budget Prediction Model
This model predicts the top 10 monthly expense categories based on historical user transactions using time-series trend analysis. It aggregates past spending by category, learns the monthly spending pattern using a Linear Regression model, and forecasts the expected amount for the upcoming month. The output helps users anticipate future expenses and plan budgets more effectively by identifying where most of their money is likely to be spent.
We collect transactions from multiple integrated data sources—including direct personal investigation, controlled experimentation (seeded/synthetic data), user-entered records, and secondary sources such as the Plaid Sandbox/API. All collected transactions are unified into a single structured dataset to ensure consistency and completeness for analysis and model training. The consolidated dataset is then exported to CSV format, enabling seamless preprocessing, visualization, and machine learning model development.


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# === Configure your input path here ===
CSV_PATH = Path("data/expenses_data.csv")  # change if needed
TOP_N = 10  # how many top categories to plot
OUT_DIR = Path("eda_outputs")

pd.set_option("display.max_rows", 20)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)

```

```python
# %% Load & clean
df_raw = pd.read_csv(CSV_PATH)
df = df_raw.copy()

# Robust datetime parsing: normalize mixed offsets to UTC then drop tz
dt = pd.to_datetime(df["transactionDate"], errors="coerce", utc=True)
bad_dates = dt.isna().sum()
if bad_dates:
    print(f"[WARN] Dropping {bad_dates} rows with unparseable dates.")

# make timezone-naive for grouping by month
df = df[~dt.isna()].copy()
df["transactionDate"] = dt[~dt.isna()].dt.tz_convert("UTC").dt.tz_localize(None)

# Keep expenses only (exclude income)
df["is_income"] = df["is_income"].astype(str).str.lower()
df = df[df["is_income"].isin(["f", "false", "0"])]

# Ensure numeric amount
df["amount"] = pd.to_numeric(df["amount"], errors="coerce")
bad_amt = df["amount"].isna().sum()
if bad_amt:
    print(f"[WARN] Dropping {bad_amt} rows with non-numeric amount.")
df = df.dropna(subset=["amount"])

print(df.dtypes)
df.head()
```

```python
# %% 1) Monthly Spending Trends Analysis — table
df["year_month"] = df["transactionDate"].dt.to_period("M")
monthly = df.groupby("year_month", as_index=False)["amount"].sum()
monthly["year_month"] = monthly["year_month"].astype(str)
monthly.head(12)
```

```python
# %% 2) Category-Wise Expense Distribution — table
# Sum by primary category bucket (cat_name)
cat_totals = df.groupby("cat_name", as_index=False)["amount"].sum().sort_values("amount", ascending=False)
cat_top = cat_totals.head(TOP_N).reset_index(drop=True)
cat_top
```

```python
# %% 2) Category-Wise Expense Distribution — plot
plt.figure()
plt.bar(cat_top["cat_name"].astype(str), cat_top["amount"])
plt.title(f"Top {TOP_N} Categories by Total Spend")
plt.xlabel("Category")
plt.ylabel("Total Amount")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# %% 3) Correlation (numeric-only)
num_cols = df.select_dtypes(include=[np.number]).copy()
corr = num_cols.corr(numeric_only=True)
corr
```

```python
# %% 3) Outlier Detection (Z-score within each category)
def zscore(x: pd.Series) -> pd.Series:
    mu = x.mean()
    sigma = x.std(ddof=0)
    if sigma == 0 or np.isnan(sigma):
        return pd.Series([0.0] * len(x), index=x.index)
    return (x - mu) / sigma

df["z_amount_by_cat"] = df.groupby("cat_name")["amount"].transform(zscore)

# Flag outliers beyond |z| >= 3
outliers = df[np.abs(df["z_amount_by_cat"]) >= 3].copy()
print(f"Outliers detected: {len(outliers)}")
outliers[["transactionDate", "cat_name", "merchant_name", "amount", "z_amount_by_cat"]].head(20)
```

```python
# %% 3) Boxplot for top categories (distribution view)
plot_data = []
labels = []
for cname in cat_top["cat_name"]:
    vals = df.loc[df["cat_name"] == cname, "amount"].values
    if len(vals) > 0:
        plot_data.append(vals)
        labels.append(str(cname))

if plot_data:
    plt.figure()
    plt.boxplot(plot_data, vert=True, labels=labels, showfliers=True)
    plt.title("Amount Distribution for Top Categories")
    plt.xlabel("Category")
    plt.ylabel("Amount")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
else:
    print("[INFO] Not enough data for boxplot.")
```

```python
# %% Export key tables to CSV
OUT_DIR.mkdir(exist_ok=True)
monthly.to_csv(OUT_DIR / "monthly_totals.csv", index=False)
cat_totals.to_csv(OUT_DIR / "category_totals.csv", index=False)
outliers.to_csv(OUT_DIR / "outliers_by_category.csv", index=False)
print("Saved:", [
    str(OUT_DIR / "monthly_totals.csv"),
    str(OUT_DIR / "category_totals.csv"),
    str(OUT_DIR / "outliers_by_category.csv"),
])
```
